```python
import easyocr
```

```python
reader = easyocr.Reader(['en'],gpu=True)
```

```python

from PIL import  ImageDraw , Image

im = Image.open("D:/WappnetProjects/invoice reading with easyocr/work/data/homedepot_bill_page-0001.jpg")
```

```python
bounds = reader.readtext("D:/WappnetProjects/invoice reading with easyocr/work/data/homedepot_bill_page-0001.jpg")
bounds
```

```python
def draw_boxes(image, bounds, color='yellow', width= 2):
    draw = ImageDraw.Draw(image)
    for bound in bounds:
        p0 ,p1, p2, p3 =bound[0]
        draw.line([*p0, *p1, *p2, *p3, *p0], fill= color, width= width)
    return image

draw_boxes(im, bounds)
```
